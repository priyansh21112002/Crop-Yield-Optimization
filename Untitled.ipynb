### 1. **Linear Programming (LP)**
   - **Objective**: Maximize crop yield based on parameters.
   - **Constraints**: Used attributes as constraints that impact crop yield. 
   - **Approach**: Used simplex-based methods to solve the LP, considering cases like degenerate solutions and duality to optimize crop production.



```python
import pandas as pd
import numpy as np
from pyomo.environ import *

# Define the results dictionary to store optimized yield and land allocation for each country
results = {}
min_diversity_ratio = 0.4  # Minimum 40% allocation for each crop as part of total land

# Iterate over each country
for country, coeffs in country_coefficients.items():
    # Extract the coefficients for this country
    pesticide_coef = coeffs['pesticide_coef']
    temp_coef = coeffs['temp_coef']
    
    # Filter data for the current country to get its available crops
    country_data = data[data['Area'] == country]
    available_crops = country_data['Item'].unique()
    
    # Create a ConcreteModel for this country
    model = ConcreteModel()
    
    # Define variables for each crop based on available crops in this country
    crop_vars = {crop: Var(within=NonNegativeReals, bounds=(10, None)) for crop in available_crops}
    
    # Add each crop variable to the model
    for crop, var in crop_vars.items():
        setattr(model, f"A_{crop.replace(' ', '_')}", var)
    
    # Objective function: Maximize yield using the derived coefficients
    model.obj = Objective(
        expr=(pesticide_coef + temp_coef) * sum(crop_vars[crop] for crop in available_crops),
        sense=maximize
    )
    
    # Land constraint - ensure the total allocation is within the available land
    model.land_constraint = Constraint(expr=sum(crop_vars[crop] for crop in available_crops) <= 100000)
    
    # Pesticide constraint (example, total available is 800 tonnes)
    model.pesticide_constraint = Constraint(
        expr=sum(crop_vars[crop] * pesticide_coef for crop in available_crops) <= 800
    )
    
    # Diversity constraint: at least a certain percentage for each crop
    for crop in available_crops:
        model.add_component(
            f"diversity_constraint_{crop}",
            Constraint(expr=crop_vars[crop] >= min_diversity_ratio * 100000 / len(available_crops))
        )
    
    # Solve the model
    solver = SolverFactory('glpk')
    result = solver.solve(model, tee=False)
    
    # Store the optimized land allocations and yield for this country
    if result.solver.status == 'ok' and result.solver.termination_condition == 'optimal':
        land_allocation = {crop: crop_vars[crop].value for crop in available_crops}
        total_yield = (pesticide_coef + temp_coef) * sum(land_allocation[crop] for crop in available_crops)
        
        results[country] = {
            "land_allocation": land_allocation,
            "total_yield": total_yield
        }

# Print the results for each country
for country, result in results.items():
    print(f"\nCountry: {country}")
    print("Optimal Land Allocation (in hectares):")
    for crop, area in result["land_allocation"].items():
        print(f"{crop}: {area:.2f} ha")
    print("\nTotal Maximized Crop Yield:")
    print(f"{result['total_yield']:.2f} hg/ha")
```

### 2. **Non-Linear Programming (NLP)**
   - **Objective**: Since crop yield often has a non-linear relationship with parameters, we  built a non-linear model that predicts yield.
   - **Constraints**: Constraints includes bounds on parameters
   - **Approach**: Implemented first and second-order conditions, line search methods, or Hessian-based algorithms (e.g., Newton's method) for non-linear optimization.

```python
import numpy as np

# Define parameters
total_land = 100000  # Total available land (in hectares)
min_diversity_ratio = 0.4  # Minimum diversity ratio for each crop
tolerance = 1e-6
max_iters = 1000

# Extract crop names per country
country_crop_dict = data.groupby("Area")["Item"].unique().to_dict()

# Define non-linear yield function based on historical data
def yield_function(pesticide_coef, temp_coef, crop_areas):
    return sum((pesticide_coef * area**0.5 + temp_coef * area**0.3) for area in crop_areas)

# Gradient of the yield function
def yield_gradient(pesticide_coef, temp_coef, crop_areas):
    return np.array([0.5 * pesticide_coef * area**(-0.5) + 0.3 * temp_coef * area**(-0.7) for area in crop_areas])

# Hessian (second derivative matrix) for the yield function with regularization
def yield_hessian(pesticide_coef, temp_coef, crop_areas, regularization_factor=1e-5):
    diag_elements = [-0.25 * pesticide_coef * area**(-1.5) - 0.21 * temp_coef * area**(-1.7) for area in crop_areas]
    return np.diag(diag_elements) + np.eye(len(crop_areas)) * regularization_factor

# Line search to find optimal step size
def line_search(crop_areas, grad, pesticide_coef, temp_coef):
    alpha = 1.0
    beta = 0.8  # Step size reduction factor
    while yield_function(pesticide_coef, temp_coef, crop_areas - alpha * grad) < yield_function(pesticide_coef, temp_coef, crop_areas):
        alpha *= beta
    return alpha

# Optimization function for each country with constraints
def optimize_yield_for_country(country, pesticide_coef, temp_coef):
    # Get the crops for the specific country
    crops = country_crop_dict[country]
    num_crops = len(crops)
    
    # Initialize crop areas respecting the minimum diversity ratio
    crop_areas = np.full(num_crops, total_land / num_crops * min_diversity_ratio)
    
    # Enforce total land constraint
    crop_areas *= total_land / crop_areas.sum()

    for _ in range(max_iters):
        # Apply gradient and Hessian updates
        grad = yield_gradient(pesticide_coef, temp_coef, crop_areas)
        hessian = yield_hessian(pesticide_coef, temp_coef, crop_areas)
        
        # Newton update using gradient and Hessian
        try:
            step_dir = np.linalg.solve(hessian, grad)
        except np.linalg.LinAlgError:
            # Handle singular matrix case by skipping update
            print(f"Singular matrix detected for {country}. Skipping update.")
            continue
        
        # Line search for optimal step size
        alpha = line_search(crop_areas, step_dir, pesticide_coef, temp_coef)
        
        # Update crop areas
        crop_areas -= alpha * step_dir
        
        # Apply constraints:
        # Land constraint: total crop areas should not exceed the available land
        if crop_areas.sum() > total_land:
            crop_areas *= total_land / crop_areas.sum()
        
        # Pesticide constraint: ensure total pesticide usage is within the limit
        total_pesticide_usage = sum(crop_areas * pesticide_coef)
        if total_pesticide_usage > 800:
            crop_areas *= 800 / total_pesticide_usage
        
        # Diversity constraint: ensure each crop area is above the minimum diversity ratio
        for i in range(num_crops):
            if crop_areas[i] < min_diversity_ratio * total_land / num_crops:
                crop_areas[i] = min_diversity_ratio * total_land / num_crops

        # Check for convergence
        if np.linalg.norm(alpha * step_dir) < tolerance:
            break
    
    return dict(zip(crops, crop_areas)), yield_function(pesticide_coef, temp_coef, crop_areas)

# Run optimization for each country
results = {}
for country, coeffs in country_coefficients.items():
    pesticide_coef = coeffs['pesticide_coef'] * 1000
    temp_coef = coeffs['temp_coef'] * 1000
    optimal_areas, max_yield = optimize_yield_for_country(country, pesticide_coef, temp_coef)
    
    results[country] = {
        "land_allocation": optimal_areas,
        "total_yield": max_yield
    }

# Print results in the desired format
for country, result in results.items():
    print(f"\nCountry: {country}")
    print("Optimal Land Allocation (in hectares):")
    for crop, area in result["land_allocation"].items():
        print(f"{crop}: {area:.2f} ha")
    print("\nTotal Maximized Crop Yield:")
    print(f"{result['total_yield']:.2f} hg/ha")
```

### 3.Quadratic Programming (QP)
   - **Objective**: Formulate a QP where you minimize the cost of soil amendments while maximizing yield based on quadratic relationships among soil parameters.
   - **Approach**: You could explore using sequential quadratic programming methods to solve this, particularly useful if your model involves complex interactions between soil parameters.

```python
import pandas as pd
import numpy as np
from gurobipy import Model, GRB, QuadExpr

# Define the results dictionary to store optimized yield and land allocation for each country
results = {}
min_diversity_ratio = 0.2  # Set a higher minimum for each crop to encourage diversity
max_crop_ratio = 0.3       # Lower the max allocation per crop for better distribution
reg_factor = 1e-3          # Increase the regularization factor for more balanced allocation

# Iterate over each country
for country, coeffs in country_coefficients.items():
    # Extract and scale the coefficients for better numerical stability
    pesticide_coef = coeffs['pesticide_coef'] / 100
    temp_coef = coeffs['temp_coef'] / 100

    # Filter data for the current country to get its available crops
    country_data = data[data['Area'] == country]
    available_crops = country_data['Item'].unique()

    # Create a Gurobi model for this country
    model = Model(country)
    model.setParam('OutputFlag', 0)  # Suppress solver output

    # Define variables for each crop based on available crops in this country
    crop_vars = {crop: model.addVar(lb=10, name=f"A_{crop.replace(' ', '_')}") for crop in available_crops}
    model.update()

    # Define the quadratic objective function with increased regularization to balance allocation
    objective = QuadExpr()
    for crop in available_crops:
        crop_var = crop_vars[crop]
        # Scale down the quadratic term and add regularization
        objective += 0.5 * (pesticide_coef + temp_coef) * crop_var ** 2 + reg_factor * crop_var

    model.setObjective(objective, GRB.MAXIMIZE)

    # Land constraint - ensure the total allocation is within the available land
    model.addConstr(sum(crop_vars[crop] for crop in available_crops) <= 100000, name="LandConstraint")

    # Pesticide constraint: ensure total pesticide usage is within the limit of 800 tonnes
    model.addConstr(sum(crop_vars[crop] for crop in available_crops) * pesticide_coef <= 800, name="PesticideConstraint")

    # Diversity constraint: enforce a minimum allocation for each crop
    for crop in available_crops:
        min_alloc = min_diversity_ratio * 10000 / len(available_crops)
        model.addConstr(crop_vars[crop] >= min_alloc, name=f"DiversityConstraint_{crop}")

    # Additional constraint to prevent over-concentration of land in a single crop
    for crop in available_crops:
        max_alloc = max_crop_ratio * 10000
        model.addConstr(crop_vars[crop] <= max_alloc, name=f"MaxCropConstraint_{crop}")

    # Optimize the model
    model.optimize()

    # Store the optimized land allocations and yield for this country if optimal solution is found
    if model.status == GRB.OPTIMAL:
        land_allocation = {crop: crop_vars[crop].x for crop in available_crops}
        total_yield = sum(0.5 * (pesticide_coef + temp_coef) * (crop_vars[crop].x ** 2) for crop in available_crops)

        results[country] = {
            "land_allocation": land_allocation,
            "total_yield": total_yield
        }

# Print the results for each country
for country, result in results.items():
    print(f"\nCountry: {country}")
    print("Optimal Land Allocation (in hectares):")
    for crop, area in result["land_allocation"].items():
        print(f"{crop}: {area:.2f} ha")
    print("\nTotal Maximized Crop Yield:")
    print(f"{result['total_yield']:.2f} hg/ha")

```

### 5. **Interior Point Methods and Trust-Region Methods**
   - **Objective**: These methods can be useful for solving large-scale or particularly complex optimization problems, such as maximizing yield while minimizing environmental impacts.
   - **Approach**: Implement interior point or trust-region methods to explore alternative solution paths in cases where other methods converge slowly or struggle with constraints.

```python
import pandas as pd
import numpy as np
from gurobipy import Model, GRB, QuadExpr

# Define the results dictionary to store optimized yield and land allocation for each country
results = {}
min_diversity_ratio = 0.4  # Minimum for crop diversity
max_crop_ratio = 0.3       # Maximum allocation per crop
reg_factor = 1e-3          # Regularization factor for balance
trust_region_radius = 5000 # Radius to limit land allocation shifts for each crop

# Iterate over each country
for country, coeffs in country_coefficients.items():
    # Extract and scale the coefficients for numerical stability
    pesticide_coef = coeffs['pesticide_coef'] / 100
    temp_coef = coeffs['temp_coef'] / 100

    # Filter data for the current country to get available crops
    country_data = data[data['Area'] == country]
    available_crops = country_data['Item'].unique()

    # Create a Gurobi model for this country
    model = Model(country)
    model.setParam('OutputFlag', 0)  # Suppress solver output

    # Define variables for each crop with bounds
    # Initialize variables within a trust region around 10,000 hectares as a starting point
    crop_vars = {
        crop: model.addVar(
            lb=max(10, 1000 - trust_region_radius),
            ub=10000 + trust_region_radius,
            name=f"A_{crop.replace(' ', '_')}"
        )
        for crop in available_crops
    }
    model.update()

    # Define the objective with regularization
    objective = QuadExpr()
    for crop in available_crops:
        crop_var = crop_vars[crop]
        objective += 0.5 * (pesticide_coef + temp_coef) * crop_var ** 2 + reg_factor * crop_var

    # Set objective function for maximization
    model.setObjective(objective, GRB.MAXIMIZE)

    # Define constraints to satisfy primal feasibility (KKT conditions)
    # Land constraint
    land_constraint = model.addConstr(
        sum(crop_vars[crop] for crop in available_crops) <= 100000,
        name="LandConstraint"
    )

    # Pesticide constraint
    pesticide_constraint = model.addConstr(
        sum(crop_vars[crop] for crop in available_crops) * pesticide_coef <= 800,
        name="PesticideConstraint"
    )

    # Diversity and concentration constraints
    for crop in available_crops:
        min_alloc = min_diversity_ratio * 10000 / len(available_crops)
        max_alloc = max_crop_ratio * 10000
        model.addConstr(crop_vars[crop] >= min_alloc, name=f"DiversityConstraint_{crop}")
        model.addConstr(crop_vars[crop] <= max_alloc, name=f"MaxCropConstraint_{crop}")

    # Optimize the model using interior point and trust-region-like constraints
    model.optimize()

    # If the solution is optimal, store results and check constraints
    if model.status == GRB.OPTIMAL:
        land_allocation = {crop: crop_vars[crop].x for crop in available_crops}
        total_yield = sum(0.5 * (pesticide_coef + temp_coef) * (crop_vars[crop].x ** 2) for crop in available_crops)

        # Check for any constraint violations (KKT-like check)
        slack_violations = []
        for constr in model.getConstrs():
            slack = model.getAttr("Slack", [constr])[0]
            if slack > 1e-6:  # Slack close to zero suggests an active constraint
                slack_violations.append((constr.ConstrName, slack))

        # Store results
        results[country] = {
            "land_allocation": land_allocation,
            "total_yield": total_yield,
            "slack_violations": slack_violations
        }

# Display results with approximate KKT checks
for country, result in results.items():
    print(f"\nCountry: {country}")
    print("Optimal Land Allocation (in hectares):")
    for crop, area in result["land_allocation"].items():
        print(f"{crop}: {area:.2f} ha")
    print("\nTotal Maximized Crop Yield:")
    print(f"{result['total_yield']:.2f} hg/ha")
    
    # Display any slack violations found
    if result["slack_violations"]:
        print("\nKKT Slack Violations (approximate):")
        for violation in result["slack_violations"]:
            print(f"Constraint {violation[0]} - Slack: {violation[1]:.4f}")
    else:
        print("\nKKT conditions satisfied.")
```
